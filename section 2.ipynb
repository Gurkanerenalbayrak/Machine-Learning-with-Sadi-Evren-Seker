```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

#veri ön işleme 
veriler = pd.read_csv('eksikveriler.csv')
veriler
```

# missing values eksik veriler 

```python

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

yas = veriler.iloc[:,1:4].values

imputer = imputer.fit(yas[:,1:4])
yas[:,1:4] = imputer.transform(yas[:,1:4])
print(yas)
```

# kategorik kolonun dönüşümü için uğraşıyoruz



```python
#encoder, nominalden veya ordinalden -> numeric verilere dönüşüyor 
ulke = veriler.iloc[:,0:1].values
from sklearn import preprocessing

le = preprocessing.LabelEncoder() # bir önceki derste fit ve transformu ayrı ayrı öğrendik, şimdi tek seferde uygulayacak 
ulke[:,0] = le.fit_transform(veriler.iloc[:,0]) # bu dataframe olarak geldi. 

print(ulke)
veriler[['ulke']]
```

```python
#one hat encoding 

# kolon başlıklarında etiketleri taşımak, her etiket için 0,1 sayısını belirlemek 
ohe = preprocessing.OneHotEncoder() #önişlemedeki öğrenme süreci 
ulke = ohe.fit_transform(ulke).toarray()
print(ulke)
```

# verinin birleşimi ve df oluşturma
