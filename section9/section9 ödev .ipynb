```python
#1.kutuphaneler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score

# veri yukleme
veriler = pd.read_csv('maaslar_yeni.csv')
veriler
#2.veri onisleme
v1 = veriler.iloc[:,2:5] # unvanseviyesi-kıdem-puan
v2_maas = veriler.iloc[:,-1:] # maaş 
V1 = v1.values
V2_maas = v2_maas.values

#------------------------gurkan---------------------------kubos----------------------

a=veriler.iloc[:,2:5] #eğitimseviyesi
b=veriler.iloc[:,-1:] #maas

A = a.values
B = b.values



#linear regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(A,B)

#lineer regresyonun p-values
import statsmodels.api as sm 
model = sm.OLS(lin_reg.predict(A),A)#burada yapmaya çalıştığımız lineer regresyonun P-Valuesini merak etmek, section4 deki gibi genel bir OLS yapmıyoruz
print(model.fit().summary())
#ols => ordinary least square




#polynomial regression

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
a_poly = poly_reg.fit_transform(A)
print(a_poly)
lin_reg2 = LinearRegression()
lin_reg2.fit(a_poly,b)

print('POLY OLS')
model2 = sm.OLS(lin_reg2.predict(poly_reg.fit_transform(A)),A)
print(model2.fit().summary())
#print('Polynomial R2 degeri')
print(r2_score(B, lin_reg2.predict(poly_reg.fit_transform(A))))




#verilerin olceklenmesi
from sklearn.preprocessing import StandardScaler

sc1=StandardScaler()
a_olcekli = sc1.fit_transform(A)

sc2=StandardScaler()
b_olcekli = np.ravel(sc2.fit_transform(B.reshape(-1,1)))

from sklearn.svm import SVR
svr_reg = SVR(kernel='rbf')
svr_reg.fit(a_olcekli,b_olcekli)


print('SVR ols Değeri')
model3 = sm.OLS(svr_reg.predict(a_olcekli),a_olcekli) # a_ölcekli dememizin sebebi SVR ın standartscaler ile birlikte değiştirilmesi lazım 
print(model3.fit().summary())

print('SVR R2 degeri')
print(r2_score(b_olcekli, svr_reg.predict(a_olcekli)))





#Decision Tree Regresyon
from sklearn.tree import DecisionTreeRegressor
r_dt = DecisionTreeRegressor(random_state=0)
r_dt.fit(A,B)


print('Decision Tree R2 degeri')
print(r2_score(B, r_dt.predict(A)))

print('decision tree ols Değeri')
model4 = sm.OLS(r_dt.predict(A),A)
print(model4.fit().summary())

          
                



#Random Forest Regresyonu
from sklearn.ensemble import RandomForestRegressor
rf_reg=RandomForestRegressor(n_estimators = 10,random_state=0)
rf_reg.fit(A,B.ravel())

print('Random Forest ols Değeri')
model5 = sm.OLS(rf_reg.predict(A),A)
print(model5.fit().summary())

print('Random Forest R2 degeri')
print(r2_score(B,rf_reg.predict(A)))



print(veriler.corr()) #korelasyon matrisi ! 







'''
#Ozet R2 değerleri
print('-----------------------')
print('Linear R2 degeri')
print(r2_score(Y, lin_reg.predict(X)))

print('Polynomial R2 degeri')
print(r2_score(Y, lin_reg2.predict(poly_reg.fit_transform(X))))

print('SVR R2 degeri')
print(r2_score(y_olcekli, svr_reg.predict(x_olcekli)))


print('Decision Tree R2 degeri')
print(r2_score(Y, r_dt.predict(X)))

print('Random Forest R2 degeri')
print(r2_score(Y, rf_reg.predict(X)))
'''
```
